```python
# Use this cell to begin your analysis, and add as many as you would like!
import requests
import nltk
from bs4 import BeautifulSoup
from collections import Counter
```

```python
# extracting HTML with Request
url = "https://www.gutenberg.org/files/16/16-h/16-h.htm"
url = requests.get(url)

url.encoding = "utf-8"
html = url.text
```

```python
soup = BeautifulSoup(html)
soup_txt = soup.get_text()
```

```python
# removing punctuation, whitespace, common stop words
tokenizer = nltk.tokenize.RegexpTokenizer(r'\w+')
tokens = tokenizer.tokenize(soup_txt)

words = []
for word in tokens:
    words.append(word.lower())
   
stop_words = nltk.corpus.stopwords.words("english")
uptd_words = [word for word in words if word not in stop_words]
```

```python
count = Counter(uptd_words)
top_ten = count.most_common(10)
print(top_ten)
```
